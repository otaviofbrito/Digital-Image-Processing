```python
import utils
import matplotlib.pyplot as plot
import matplotlib.image as image
import math
import numpy as np
```

```python
plot.imshow(img.matrix, cmap='gray')
plot.axis('off') 
plot.show()
#76
```

```python
histograma = utils.histogram(img)

plot.figure()
plot.title("Histograma da imagem em tons de cinza")
plot.xlabel("Valores dos pixels")
plot.ylabel("Nmero de pixels")
plot.plot(histograma, color='black')
plot.xlim([0, 255])
plot.show()
```

### STEP 1. Primeira Limiarizao 



```python
gl = img.gray_level
#Choose an appropriate threshold
lim = 100
xpoints = np.array([0, lim, lim, gl])
ypoints = np.array([0, 0, gl, gl])
plot.plot(xpoints, ypoints)
plot.title("Arbitrary thresholding function") 
plot.show()
```

```python
utils.threshold(img,lim)
```

```python
plot.imshow(img.matrix, cmap='binary')
plot.axis('off') 
plot.title("Result of thresholding") 
```
